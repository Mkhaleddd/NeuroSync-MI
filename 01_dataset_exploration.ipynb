```python
import torch.nn as nn
import torch.nn.functional as F

class EEGNet(nn.Module):
    def __init__(self, n_channels, n_timepoints, n_classes):
        super(EEGNet, self).__init__()
        self.firstconv = nn.Sequential(
            nn.Conv2d(1, 16, (1, 3), padding=(0,1)),  # temporal conv
            nn.BatchNorm2d(16)
        )
        self.depthwise = nn.Sequential(
            nn.Conv2d(16, 32, (n_channels,1), groups=16),  # depthwise
            nn.BatchNorm2d(32),
            nn.ELU(),
            nn.AvgPool2d((1,4)),
            nn.Dropout(0.3)
        )
        self.separable = nn.Sequential(
            nn.Conv2d(32, 32, (1, 3), padding=(0,1)),  # separable conv
            nn.BatchNorm2d(32),
            nn.ELU(),
            nn.AvgPool2d((1,4)),
            nn.Dropout(0.3)
        )
        # Compute flattened size dynamically
        dummy = torch.zeros(1,1,n_channels,n_timepoints)
        dummy = self.separable(self.depthwise(self.firstconv(dummy)))
        self.fc = nn.Linear(dummy.shape[1]*dummy.shape[2]*dummy.shape[3], n_classes)

    def forward(self, x):
        x = self.firstconv(x)
        x = self.depthwise(x)
        x = self.separable(x)
        x = x.view(x.size(0), -1)
        return self.fc(x)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import freqz, butter 

fs = 250
low = 8
high = 30
order = 4 

nyq = 0.5 * fs
low_norm = low / nyq
high_norm = high / nyq


b, a = butter(order, [low_norm, high_norm], btype='band')

w, h = freqz(b, a, worN=8000)


frequencies = w * fs / (2 * np.pi)


plt.figure(figsize=(8,4))
plt.plot(frequencies, np.abs(h), 'b', label="Filter Response")
plt.axvspan(low, high, color='red', alpha=0.1, label='Passband') 
plt.title("Butterworth Bandpass Filter Frequency Response")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Gain")
plt.xlim(0, 60)
plt.grid(True)
plt.legend()
plt.show()
```

```python
from scipy.signal import butter, filtfilt

def bandpass_filter(data, low=8, high=30, fs=250, order=8):
    nyq = 0.5 * fs
    low = low / nyq
    high = high / nyq
    b, a = butter(order, [low, high], btype='band')
    return filtfilt(b, a, data, axis=-1)

X_filtered = bandpass_filter(X)
trial = 0
channel = 4  

signal_original = X[trial, channel]
signal_filtered = X_filtered[trial, channel]

def compute_fft(signal, fs):
    fft_vals = np.fft.rfft(signal)
    fft_freq = np.fft.rfftfreq(len(signal), 1/fs)
    return fft_freq, np.abs(fft_vals)

freq_orig, fft_orig = compute_fft(signal_original, fs)
freq_filt, fft_filt = compute_fft(signal_filtered, fs)
```

```python
plt.figure(figsize=(10,5))

plt.plot(freq_orig, fft_orig, label="Original")
plt.plot(freq_filt, fft_filt, label="Filtered")

plt.xlim(0, 60)
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.title("FFT Before vs After Filtering")
plt.legend()
plt.grid()
plt.show()
```

```python
class EEGSpectrumNet(nn.Module):
    def __init__(self, n_classes=2):
        super().__init__()
        
        self.features = nn.Sequential(
            nn.Conv2d(1, 32, kernel_size=(3, 3), padding=(1, 1)),
            nn.BatchNorm2d(32),
            nn.ELU(),
            nn.MaxPool2d(kernel_size=(1, 2)), 
            nn.Conv2d(32, 64, kernel_size=(3, 3), padding=(1, 1)),
            nn.BatchNorm2d(64),
            nn.ELU(),
            nn.AdaptiveAvgPool2d((1, 1)) 
        )
        
        self.classifier = nn.Linear(64, n_classes)

    def forward(self, x):
        x = self.features(x)
        x = x.view(x.size(0), -1) 
        return self.classifier(x)
```

```python
import torch.optim as optim
from torch.utils.data import DataLoader, Subset
from sklearn.model_selection import KFold
from sklearn.metrics import confusion_matrix, classification_report


device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
kf = KFold(n_splits=5, shuffle=True, random_state=42)

fold_results = []

for fold, (train_idx, val_idx) in enumerate(kf.split(X_tensor)):
    print(f"\n--- Fold {fold+1} ---")
    
    train_subset = Subset(dataset, train_idx)
    val_subset = Subset(dataset, val_idx)
    
    train_loader = DataLoader(train_subset, batch_size=32, shuffle=True)
    val_loader = DataLoader(val_subset, batch_size=32, shuffle=False)
    
    model = EEGSpectrumNet().to(device)
    optimizer = optim.Adam(model.parameters(), lr=0.0001)
    criterion = nn.CrossEntropyLoss()

    epochs = 10
    model.train()
    for epoch in range(epochs):
        for X_batch, y_batch in train_loader:
            X_batch, y_batch = X_batch.to(device), y_batch.to(device)
            optimizer.zero_grad()
            output = model(X_batch)
            loss = criterion(output, y_batch)
            loss.backward()
            optimizer.step()
    

    model.eval()
    y_true = []
    y_pred = []
    with torch.no_grad():
        for X_batch, y_batch in val_loader:
            X_batch, y_batch = X_batch.to(device), y_batch.to(device)
            output = model(X_batch)
            preds = output.argmax(dim=1)
            y_true.extend(y_batch.cpu().numpy())
            y_pred.extend(preds.cpu().numpy())
    
    acc = np.mean(np.array(y_true) == np.array(y_pred))
    print(f"Fold {fold+1} Accuracy: {acc:.4f}")
    fold_results.append(acc)

    cm = confusion_matrix(y_true, y_pred)
    print("Confusion Matrix:\n", cm)
    print("Classification Report:\n", classification_report(y_true, y_pred))

print(f"\nAverage Accuracy: {np.mean(fold_results):.4f}")
```

```python
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset, Subset
from scipy.signal import butter, filtfilt, stft
from sklearn.model_selection import KFold
from sklearn.metrics import confusion_matrix, classification_report

def get_clean_spectrogram(X, fs=250):
    nyq = 0.5 * fs
    b, a = butter(4, [8/nyq, 30/nyq], btype='band')
    X_filt = filtfilt(b, a, X, axis=-1)

    f, t, Zxx = stft(X_filt, fs=fs, nperseg=64, noverlap=32, axis=-1)
    X_spec = np.abs(Zxx)
    X_spec = np.log1p(X_spec) 
   
    X_spec = (X_spec - np.mean(X_spec)) / (np.std(X_spec) + 1e-6)
    return X_spec.astype('float32')

class EEGSpectrumNet(nn.Module):
    def __init__(self, n_classes=2):
        super().__init__()
        self.conv_block = nn.Sequential(
            nn.Conv2d(1, 16, kernel_size=3, padding=1),
            nn.BatchNorm2d(16),
            nn.ELU(),
            nn.Conv2d(16, 32, kernel_size=3, padding=1),
            nn.BatchNorm2d(32),
            nn.ELU(),
            nn.AdaptiveAvgPool2d((1, 1)) 
        )
        self.classifier = nn.Linear(32, n_classes)

    def forward(self, x):
        x = self.conv_block(x)
        x = x.view(x.size(0), -1) 
        return self.classifier(x)


print("Processing Spectrograms...")
X_spec = get_clean_spectrogram(X)
trials, n_channels, f_bins, t_bins = X_spec.shape

X_tensor = torch.tensor(X_spec, dtype=torch.float32).reshape(trials, 1, n_channels * f_bins, t_bins)
y_tensor = torch.tensor(y_numeric, dtype=torch.long).view(-1)

print(f"Data Ready! Tensor Shape: {X_tensor.shape}") 

dataset = TensorDataset(X_tensor, y_tensor)

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
kf = KFold(n_splits=9, shuffle=True, random_state=42)
fold_results = []

for fold, (train_idx, val_idx) in enumerate(kf.split(X_tensor)):
    print(f"\n--- Fold {fold+1} ---")
    
    train_loader = DataLoader(Subset(dataset, train_idx), batch_size=16, shuffle=True)
    val_loader = DataLoader(Subset(dataset, val_idx), batch_size=16)

    model = EEGSpectrumNet().to(device)
    optimizer = optim.Adam(model.parameters(), lr=0.0005, weight_decay=1e-3)
    criterion = nn.CrossEntropyLoss()

    epochs = 50 
    model.train()
    for epoch in range(epochs):
        for batch_X, batch_y in train_loader:
            batch_X, batch_y = batch_X.to(device), batch_y.to(device)
            optimizer.zero_grad()
            loss = criterion(model(batch_X), batch_y)
            loss.backward()
            optimizer.step()

    model.eval()
    y_true, y_pred = [], []
    with torch.no_grad():
        for batch_X, batch_y in val_loader:
            batch_X, batch_y = batch_X.to(device), batch_y.to(device)
            preds = model(batch_X).argmax(dim=1)
            y_true.extend(batch_y.cpu().numpy())
            y_pred.extend(preds.cpu().numpy())
    
    acc = np.mean(np.array(y_true) == np.array(y_pred))
    fold_results.append(acc)
    print(f"Accuracy: {acc:.4f}")

print(f"\nFinal Average Accuracy: {np.mean(fold_results):.4f}")
```
